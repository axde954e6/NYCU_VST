# split result

## Original YOLOv4

```python
file_path="/content/gdrive/MyDrive/VSAT_HW/darknet/result.txt"
result_file=open(file_path)
save_path="/content/gdrive/MyDrive/VSAT_HW/Original/"
now=0
for line in result_file.readlines():
  if(now<13):
    now+=1
    continue
  else:
    if line.find("data/test_dataset/")!=-1:
      img_name=line.lstrip("data/test_dataset/")
      img_name=img_name.split(".jpg")[0]
      save_file=open(save_path+img_name+'.txt','w')
    elif line.find("Enter")!=-1:
      save_file.close()
    elif line.find("car")!=-1:
      
      line=line[line.find(": ")+2:]
      if line.split("%")[0]=="100":
        confid="1"
      else:
        confid="0."+line.split("%")[0]
      line=line.split('left_x:')[1]
      line=line.lstrip(" ")
      left_x=line.split(" ")[0]
      line=line.split("top_y:")[1]
      line=line.lstrip(" ")
      top_y=line.split(" ")[0]
      line=line.split("width:")[1]
      line=line.lstrip(" ")
      width=line.split(" ")[0]
      line=line.split("height:")[1]
      line=line.lstrip(" ")
      height=line.split(")")[0]
      right_x=str(int(left_x)+int(width))
      bottom_y=str(int(top_y)+int(height))
      to_write="0 "+confid+" "+left_x+" "+top_y+" "+right_x+" "+bottom_y+"\n"
      save_file.write(to_write)
```

## YOLOv4 with SE module

```python
import os
get_path='/content/drive/MyDrive/VSAT_HW/PyTorch_YOLOv4/runs/test/exp8/labels/'
save_path='/content/drive/MyDrive/VSAT_HW/Modified/'
for item in os.listdir(get_path):
  get_file=open(get_path+item,'r')
  save_file=open(save_path+item,'w')
  for line in get_file.readlines():
    get_list=line.split()
    conf=float(get_list[1])
    if conf<0.2:
      break
    left=int((float(get_list[2])-float(get_list[4])/2)*1920)
    right=int((float(get_list[2])+float(get_list[4])/2)*1920)
    top=int((float(get_list[3])-float(get_list[5])/2)*1080)
    down=int((float(get_list[3])+float(get_list[5])/2)*1080)
    to_write="0 {} {} {} {} {}\n".format(conf,left,top,right,down)
    save_file.write(to_write)
  get_file.close()
  save_file.close()
```
